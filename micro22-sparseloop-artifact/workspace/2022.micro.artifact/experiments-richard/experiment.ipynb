```python
import numpy as np
import matplotlib.pyplot as plt

def plot_energy_stacked_by_experiment(experiments_paths):
    """
    Stack the per-level energy for several Timeloop runs.

    Parameters
    ----------
    experiments_paths : list[str]
        Paths to the *.map+stats.xml* files produced by Timeloop.
        Each path is treated as a separate experiment.
    """
    # ---------------------------------------------------------------
    # 1.  Parse the stats from every experiment
    # ---------------------------------------------------------------
    runs   = []
    levels = None                      # to remember the level order
    for p in experiments_paths:
        out = parse_timeloop_stats(p)["energy_breakdown_pJ"]

        # Save the level order from the *first* experiment and assume
        # every other run has the same set (common for Timeloop output).
        if levels is None:
            levels = list(out.keys())

        runs.append([out[lvl]["energy"] for lvl in levels])

    runs = np.array(runs)              # shape: (n_experiments, n_levels)

    # ---------------------------------------------------------------
    # 2.  Build the stacked bars
    # ---------------------------------------------------------------
    x      = np.arange(len(experiments_paths))
    width  = 0.6
    bottom = np.zeros(len(experiments_paths))

    fig, ax = plt.subplots(figsize=(12, 6))

    for lvl_idx, lvl_name in enumerate(levels):
        ax.bar(
            x,
            runs[:, lvl_idx],
            width,
            bottom=bottom,
            label=lvl_name,
            # No explicit colour  next colour in Matplotlib cycle
        )
        bottom += runs[:, lvl_idx]

    # ---------------------------------------------------------------
    # 3.  Cosmetics
    # ---------------------------------------------------------------
    ax.set_ylabel("Energy (pJ)")
    ax.set_title("Total Energy per Experiment (stacked by memory level)")
    ax.set_xticks(x)
    ax.set_xticklabels([f"Run {i+1}" for i in x], rotation=0)
    ax.legend(title="Memory Level", bbox_to_anchor=(1.02, 1), loc="upper left")
    ax.grid(axis="y", linestyle="--", alpha=0.6)
    plt.tight_layout()
    plt.show()

# Example usage ------------------------------------------------------
stats_list = [name.strip(".yaml") + "/timeloop-model.map+stats.xml" for name in names]
plot_energy_stacked_by_experiment(stats_list)
```
